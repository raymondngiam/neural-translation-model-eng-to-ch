```python
import pandas as pd
training_hist = pd.read_json('data/training_hist.json')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, axes = plt.subplots(figsize=(12, 8))
axes.set_ylabel("Loss", fontsize=14)
axes.set_xlabel("Epoch", fontsize=14)
axes.plot(training_hist['loss'],label="training")
axes.plot(training_hist['val_loss'],label="validation")
axes.legend(loc='best')
axes.set_title("Loss vs Epoch")
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(12, 8))
axes.set_ylabel("Accuracy", fontsize=14)
axes.set_xlabel("Epoch", fontsize=14)
axes.set_ylim((0,1.0))
axes.plot(training_hist['sparse_categorical_accuracy'],label="training")
axes.plot(training_hist['val_sparse_categorical_accuracy'],label="validation")
axes.legend(loc='best')
axes.set_title("Accuracy vs Epoch")
plt.show()
```
